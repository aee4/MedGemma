```python
from transformers import pipeline
from PIL import Image
import requests
import torch



```

```python
import torch, gc
gc.collect()
torch.cuda.empty_cache()
```

```python
# Use text-to-text pipel
pipe = pipeline(
    "text-generation",
    model="google/medgemma-4b-pt",
    torch_dtype="auto",              # auto-selects fastest dtype (float16 on GPU)
    device_map="auto"
)

# Example prompt
prompt = "Explain the symptoms and causes of eczema."

# Generate concise output
output = pipe(
    prompt,
    max_new_tokens= 200,
    do_sample=False                  # disables random sampling for speed & consistency
)

print(output[0]["generated_text"])
```
