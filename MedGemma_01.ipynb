```python
# Use text-to-text pipel
pipe = pipeline(
    "text-generation",
    model="google/medgemma-4b-pt",
    torch_dtype="auto",              # auto-selects fastest dtype (float16 on GPU)
    device_map="auto"
)

# Example prompt
prompt = "Explain the symptoms and causes of eczema."

# Generate concise output
output = pipe(
    prompt,
    max_new_tokens= 200,
    do_sample=False                  # disables random sampling for speed & consistency
)

print(output[0]["generated_text"])
```

```python
from transformers import pipeline
from PIL import Image
import torch
import io

# create a blank dummy image
blank_image = Image.new("RGB", (224, 224), (255, 255, 255))

pipe = pipeline(
    "image-text-to-text",
    model="google/medgemma-4b-pt",
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

prompt = (
    "Patient report: baby scratching red bumpy patches, can’t sleep, skin dry and inflamed. "
    "What is the likely diagnosis?"
)

output = pipe(blank_image, prompt, max_new_tokens=200, do_sample=False)
print(output[0]["generated_text"])
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="google/medgemma-4b-pt",
    torch_dtype="auto",
    device_map="auto"
)

system_prompt = (
    "You are a professional dermatologist AI. "
    "You give concise, medically sound answers based on symptom descriptions."
)

user_prompt = (
    "I'm not sure if I have eczema, but I have these itchy, red patches on my skin. I want to know what I can do to manage it and prevent it from getting worse. "
    "What’s the most likely diagnosis?"
)

prompt = f"{system_prompt}\n\nUser: {user_prompt}\nAI:"
output = pipe(prompt, max_new_tokens=150, do_sample=False)

print(output[0]["generated_text"])
```
