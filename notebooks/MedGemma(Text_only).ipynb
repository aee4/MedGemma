```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="google/medgemma-4b-pt",
    torch_dtype="auto",
    device_map="auto"
)

system_prompt = (
    "You are a professional dermatologist AI.You give concise, medically sound diagnosis based on symptom descriptions."
)

user_prompt = (
    "I've developed these dry, itchy patches on my hands and arms. It's making it hard to focus at work and even do simple tasks like washing dishes."
)

prompt = f"{system_prompt}\n\nUser: {user_prompt}\nAI:"
output = pipe(prompt, max_new_tokens=120, do_sample=False)

print(output[0]["generated_text"])
```
