```python
from PIL import Image
from transformers import pipeline, AutoProcessor

model_id = "google/medgemma-4b-pt"

# Load processor and check the actual image token
processor = AutoProcessor.from_pretrained(model_id)
print("Image token(s):", processor.image_token)

# Load image
image = Image.open("EC (5) (1).jpg")

# Load pipeline
pipe = pipeline(
    "image-text-to-text",
    model=model_id,
    torch_dtype="auto",
    device_map="auto"
)

# Use the processor’s correct image token dynamically
image_token = processor.image_token if hasattr(processor, "image_token") else "<image>"
prompt = (
    f"System: You are a dermatology expert analyzing patient skin images. "
    f"User: Describe the visible skin condition and possible diagnosis. {image_token}"
)

# Generate
output = pipe(
    {"images": [image], "text": prompt},
    max_new_tokens=300,
)

print(output[0]["generated_text"])
```
